```python
# Plot distribution of log_vedba from a VEDBA file
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os

# Read one of the VEDBA files (you can change the filename)
# Extract tag from file name
vedba_file = '/mnt/EAS_shared/baboon/working/data/processed/2025/acc/vedba/24AA01_5O8B_10346.parquet'
tag = os.path.basename(vedba_file).replace('.parquet', '')
df = pd.read_parquet(vedba_file)

print(f"Loaded VEDBA data for animal: {tag}")
print(f"Number of records: {len(df):,}")
print(f"Date range: {df['timestamp'].min()} to {df['timestamp'].max()}")
print(f"\nDataset columns: {list(df.columns)}")

# Display basic statistics for log_vedba
print(f"\nLog VEDBA statistics:")
print(df['logvedba'].describe())

# Create the distribution plot
plt.figure(figsize=(12, 8))

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# 1. Histogram of log_vedba
axes[0, 0].hist(df['logvedba'], bins=50, alpha=0.7, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Distribution of Log VEDBA', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Log VEDBA')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].grid(True, alpha=0.3)

# 2. Density plot
axes[0, 1].hist(df['logvedba'], bins=50, density=True, alpha=0.7, color='lightcoral', edgecolor='black')
axes[0, 1].set_title('Density Distribution of Log VEDBA', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Log VEDBA')
axes[0, 1].set_ylabel('Density')
axes[0, 1].grid(True, alpha=0.3)

# 3. Box plot
axes[1, 0].boxplot(df['logvedba'], vert=True)
axes[1, 0].set_title('Box Plot of Log VEDBA', fontsize=14, fontweight='bold')
axes[1, 0].set_ylabel('Log VEDBA')
axes[1, 0].grid(True, alpha=0.3)

# 4. Time series plot (sample of data)
# Sample every 100th point for cleaner visualization
sample_df = df.iloc[::100].copy()
axes[1, 1].plot(sample_df['timestamp'], sample_df['logvedba'], alpha=0.7, color='green', linewidth=0.5)
axes[1, 1].set_title('Log VEDBA Over Time (Sampled)', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Time')
axes[1, 1].set_ylabel('Log VEDBA')
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Additional statistics
print(f"\nAdditional statistics:")
print(f"Mean log_vedba: {df['logvedba'].mean():.3f}")
print(f"Median log_vedba: {df['logvedba'].median():.3f}")
print(f"Standard deviation: {df['logvedba'].std():.3f}")
print(f"Skewness: {df['logvedba'].skew():.3f}")
print(f"Kurtosis: {df['logvedba'].kurtosis():.3f}")

# Show some sample data
print(f"\nFirst 5 rows of data:")
print(df.head())
```

```python
# Create summary plots for all VEDBA files
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
from pathlib import Path
from matplotlib.backends.backend_pdf import PdfPages
import seaborn as sns
from datetime import datetime
from scipy import stats

# Create output directory for plots
plots_dir = 'plots/vedba_summary'
os.makedirs(plots_dir, exist_ok=True)

# Get all VEDBA files
vedba_files = glob.glob('/mnt/EAS_shared/baboon/working/data/processed/2025/acc/vedba/*.parquet')
print(f"ðŸ” Found {len(vedba_files)} VEDBA files to analyze")
print(f"ðŸ“ Plots will be saved to: {plots_dir}/")

# Store data for all animals
all_data = []
data_collection_info = []

print(f"ðŸ“Š Loading data from all animals...")

# Load data from all files
for i, vedba_file in enumerate(vedba_files, 1):
    filename = os.path.basename(vedba_file)
    animal_id = filename.replace('.parquet', '')
    
    print(f"[{i:2d}/{len(vedba_files)}] Loading {animal_id}...")
    
    try:
        # Read the data
        df = pd.read_parquet(vedba_file)
        
        # Add animal_id for grouping
        df['animal_id'] = animal_id
        all_data.append(df[['animal_id', 'logvedba', 'timestamp']])
        
        # Create data collection info (daily presence)
        df['date'] = pd.to_datetime(df['timestamp']).dt.date
        daily_data = df.groupby('date').size().reset_index(name='records_count')
        daily_data['animal_id'] = animal_id
        data_collection_info.append(daily_data[['animal_id', 'date', 'records_count']])
        
    except Exception as e:
        print(f"    âŒ Error processing {animal_id}: {str(e)}")
        continue

# Combine all data
print(f"ðŸ”— Combining data...")
if all_data and data_collection_info:
    combined_df = pd.concat(all_data, ignore_index=True)
    collection_df = pd.concat(data_collection_info, ignore_index=True)
else:
    print("âŒ No data loaded. Skipping further analysis.")
    combined_df = pd.DataFrame()
    collection_df = pd.DataFrame()

print(f"âœ… Data loaded successfully!")
print(f"   Total records: {len(combined_df):,}")
print(f"   Animals: {combined_df['animal_id'].nunique()}")
print(f"   Date range: {combined_df['timestamp'].min()} to {combined_df['timestamp'].max()}")

# ==========================================
# PLOT 1: Line Plot Density Curves by Animal
# ==========================================

print(f"ðŸ“ˆ Creating density line plot...")

# Create the line plot
plt.style.use('default')
fig, ax = plt.subplots(figsize=(16, 12))

# Get unique animals and create colors
unique_animals = sorted(combined_df['animal_id'].unique())
colors = plt.cm.tab20(np.linspace(0, 1, len(unique_animals)))

# Create density lines for each animal
x_range = np.linspace(combined_df['logvedba'].min(), combined_df['logvedba'].max(), 200)

for i, animal in enumerate(unique_animals):
    animal_data = combined_df[combined_df['animal_id'] == animal]['logvedba']

    # Calculate kernel density estimation
    if len(animal_data) > 10:  # Only plot if enough data points
        try:
            density = stats.gaussian_kde(animal_data)
            density_values = density(x_range)
            
            # Plot density line
            ax.plot(x_range, density_values, 
                   label=animal, color=colors[i], linewidth=1.5, alpha=0.7)
        except:
            # Fallback for problematic data
            print(f"    âš ï¸  Skipping {animal} due to density calculation issues")
            continue

ax.set_xlabel('Log VEDBA', fontsize=14, fontweight='bold')
ax.set_ylabel('Density', fontsize=14, fontweight='bold')
ax.set_title('Distribution of Log VEDBA by Individual (Density Curves)', fontsize=16, fontweight='bold')
ax.grid(True, alpha=0.3)

# Add legend with multiple columns (smaller font to fit all animals)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', ncol=4, fontsize=6)

plt.tight_layout()

# Save as PDF
pdf_file1 = f"{plots_dir}/log_vedba_density_lines.pdf"
plt.savefig(pdf_file1, dpi=300, bbox_inches='tight')
print(f"ðŸ’¾ Saved density line plot: {pdf_file1}")
plt.show()

# ==========================================
# PLOT 2: Data Collection Effort
# ==========================================

print(f"ðŸ“… Creating data collection effort plot...")

# Prepare data for collection effort plot
collection_df['date'] = pd.to_datetime(collection_df['date'])

# Create a more readable plot with animal order
animals_sorted = sorted(collection_df['animal_id'].unique())

# Create the data collection effort plot
fig, ax = plt.subplots(figsize=(20, 16))

# Create y-positions for each animal
y_positions = {animal: i for i, animal in enumerate(animals_sorted)}

# Plot points for each day with data
for _, row in collection_df.iterrows():
    y_pos = y_positions[row['animal_id']]
    # Size points based on number of records (scaled)
    point_size = min(50, max(5, row['records_count'] / 100))
    ax.scatter(row['date'], y_pos, s=point_size, alpha=0.6, c='steelblue')

# Customize the plot
ax.set_xlabel('Date', fontsize=14, fontweight='bold')
ax.set_ylabel('Individual ID', fontsize=14, fontweight='bold')
ax.set_title('Data Collection Effort by Individual Over Time', fontsize=16, fontweight='bold')

# Set y-axis labels
ax.set_yticks(range(len(animals_sorted)))
ax.set_yticklabels(animals_sorted, fontsize=8)

# Format x-axis
ax.tick_params(axis='x', rotation=45)

# Add grid
ax.grid(True, alpha=0.3)

# Add colorbar legend for point sizes
# Create a dummy scatter for legend
sizes = [5, 25, 50]
labels = ['Low (<100)', 'Medium (100-500)', 'High (>500)']
legend_elements = [plt.scatter([], [], s=size, alpha=0.6, c='steelblue', label=label) 
                  for size, label in zip(sizes, labels)]
legend1 = ax.legend(handles=legend_elements, title='Records per day', 
                   loc='upper right', fontsize=10)
legend1.set_title('Records per day', prop={'weight': 'bold'})

plt.tight_layout()

# Save as PDF
pdf_file2 = f"{plots_dir}/data_collection_effort.pdf"
plt.savefig(pdf_file2, dpi=300, bbox_inches='tight')
print(f"ðŸ’¾ Saved data collection effort: {pdf_file2}")
plt.show()

# ==========================================
# SUMMARY STATISTICS
# ==========================================

print(f"ðŸ“Š Creating summary statistics...")

# Calculate summary statistics
summary_stats = combined_df.groupby('animal_id')['logvedba'].agg([
    'count', 'mean', 'std', 'min', 'max', 'median'
]).round(3).reset_index()

summary_stats.columns = ['animal_id', 'n_records', 'mean_log_vedba', 'std_logvedba', 
                        'min_logvedba', 'max_logvedba', 'median_logvedba']

# Add data collection period info
collection_summary = collection_df.groupby('animal_id')['date'].agg(['min', 'max', 'count']).reset_index()
collection_summary.columns = ['animal_id', 'first_date', 'last_date', 'days_with_data']

# Merge summaries
final_summary = summary_stats.merge(collection_summary, on='animal_id')

# Save summary
summary_file = f"{plots_dir}/vedba_analysis_summary.csv"
final_summary.to_csv(summary_file, index=False)

print(f"ðŸ’¾ Summary statistics saved: {summary_file}")

print(f"\nâœ… Analysis complete!")
print(f"ðŸ“‹ Summary across all animals:")
print(f"   Total animals: {len(final_summary)}")
print(f"   Total records: {final_summary['n_records'].sum():,}")
print(f"   Mean log_vedba: {final_summary['mean_logvedba'].mean():.3f} Â± {final_summary['mean_logvedba'].std():.3f}")
print(f"   Data collection period: {collection_df['date'].min()} to {collection_df['date'].max()}")

# Show first few animals
print(f"\nFirst 5 animals summary:")
print(final_summary[['animal_id', 'n_records', 'mean_logvedba', 'days_with_data']].head())
```
